```python
import pandas as pd
import seaborn as sns
```

```python
districts=['cumilla','Dhaka','Rangpur','Khulna','jashor']
complaint_type=['power cut','trafic','noice','sarbage','water park']
n_rows=500
data={
    'district':np.random.choice(districts,size=n_rows),
    'complaint_type':np.random.choice(complaint_type,size=n_rows),
    'temperature':np.random.normal(loc=25,scale=5,size=n_rows),
    'eletricity_bill':np.random.normal(loc=500,scale=100,size=n_rows),
    'num_ice_cream_cart':np.random.choice([1,3,4,5,10,10,10],size=n_rows),
    'survey_rating':np.random.choice([1,8,9,2,3,6,np.nan],size=n_rows)

}
```

```python
df=pd.DataFrame(data)
```

```python
df.head()
```

```python
df.shape
```

```python
df.loc[5:10,'complaint_type']=np.nan
df=pd.concat([df,df.iloc[0:5]])
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(8,4))
sns.histplot(df['temperature'],kde=True,bins=10)
plt.title('Temperature Distribution')
plt.xlabel('temperature degree')
plt.ylabel('frequency')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x='district',y='eletricity_bill',data=df)
plt.title('Electricity Bill by District')
plt.xlabel('district')
plt.ylabel('electricity bill')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='plasma')
plt.title('Correlation Heatmap')
plt.show()
```
