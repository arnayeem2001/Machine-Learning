```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({
    'Name':['Alice','Bob',np.nan,'Eve'],
    'Age':[25,30,np.nan,23],
    'City':['Dhaka',np.nan,'Cumilla',np.nan]
})
```

```python
df
```

```python
df.dropna()
```

```python
df
```

```python
df=df.fillna(
    {
        'Name':'Unknown',
        'Age':df['Age'].mean(),
        'City':"Model Town"
    }
)
```

```python
df
```

```python
df['Age']=df['Age'].astype(int)
```

```python
df
```
